# NLST Sybil FM demo

In this notebook, we demonstrate the use of foundation models (FMs) to aid in downstream tasks. We make use of Imaging Data Commons (IDC) data, which holds a large lung cancer imaging dataset, the National Lung Screening Trial (NLST), along with relevant metadata including lung cancer stage.

Here we compare two FMs, FMCIB that uses 3D volumes, and MedSigLIP that uses 2D patches. For both, we extract the embeddings, and use those to train a classifier to predict the lung cancer stage.

Here we assume that we have already created the DICOM Structured Reports (SRs) that hold the bounding box annotations from Sybil. Temporarily, we load from a bucket. Later, we will load from Zenodo, or GitHub release attachments.

Part 1 - NLST_Sybil_FM_demo_part1.ipynb
  - Inputs: SRs and metadata
  - Outputs: nlst_sybil_fm.csv
  - Code:
    - Join BigQuery tables that hold the metadata extracted from the SRs with the clincal metadata in IDC  

Part 2 - NLST_Sybil_FM_demo_part2.ipynb
  - Inputs: nlst_sybil_fm.csv
  - Outputs: fmcib_input.csv and fmcib_features.csv
  - Code:
    - Download DICOM data
    - Convert to nifti
    - Create FMCIB features csv

Part 3 - NLST_Sybil_FM_demo_part3.ipynb
  - Inputs: nlst_sybil_fm.csv
  - Outputs: medsiglip_features.csv
  - Code:
    - Download DICOM data
    - Create MedSIGLip features csv

Part 4 - NLST_Sybil_FM_demo_part4.ipynb
  - Inputs: nlst_sybil_fm.csv, fmcib_features.csv, medisglip_features.csv
  - Outputs: trained models and evaluation
  - Code:
    - Train/val/test two classification models using FMCIB and MedSIGLip features and compare
